## Menentukan Pertanyaan Bisnis

**SMART Question** adalah sebuah framework untuk merumuskan pertanyaan secara terstruktur agar memeperoleh informasi yang mendalam.

* **Spesific (Spesifik)**
  * Pertanyaan harus jelas, fokus pada sebuah topik tertentu, dan tidak bermakna ganda. Hindari pertanyaan yang terlalu luas.
    * Salah: Bagaimana cara meningkatkan penjualan?
    * Benar: Faktor apa saja yang memengaruhi penurunan penjualan produk kategori elektronik di wilayah Jakarta selama kuartal terakhir?
* **Measurable (Terukur)**
  * Pertanyaan harus bisa dijawab dengan angka atau matrix yang konkret, harus tahu apa yang akan dihitung.
    * Salah: Apakah pelanggan senang dengan layanan kita?
    * Benar: Berapa skor rata-rata Customer Satisfaction untuk layanan purna jual bulan ini dibandingkan bulan lalu?
* **Action-Oriented (Berorientasi Aksi)**
  * Hasil dari pertanyaan harus bisa memberikan arahan untuk melakukan tindakan nyata. Jika pertanyaan terjawab, stakeholder harus tahu apa langkah selanjutnya.
    * Salah: Mengapa orang suka berbelanja?
    * Benar: Fitur apa pada aplikasi yang paling sering digunakan sebelum pengguna memutuskan untuk melakukan checkout?
* **Relevant (Relevan)**
  * Hasil dari pertanyaan harus sejalan dengan tujuan utama bisnis atau masalah yang sedang dihadapi.
    * Salah: Menanyakan tentang stok gudang saat masalah utamanya adalah efektivitas kampanye media sosial.
    * Benar: Apakah kampanye iklan di Instagram memberikan Return on Ad Spend (ROAS) yang lebih tinggi dibandingkan iklan di TikTok?
* **Time-bound (Terikat Waktu)**
  * Pertanyaan harus ada batasan waktu yang jelas agar analisis memiliki konteks yang tepat.
    * Salah: Berapa banyak pengguna baru kita?
    * Benar: Berapa tingkat pertumbuhan pengguna baru secara bulanan (Month-over-Month) sepanjang tahun 2025?

## Menentukan Pertanyaan Bisnis

**1. Bagaimana pengaruh musim (season) terhadap rata-rata jumlah penyewaan sepeda harian selama tahun 2011 dan 2012?**
- **Specific (Spesifik):** Fokus secara spesifik pada pengaruh perubahan "musim" (Spring, Summer, Fall, Winter) terhadap "jumlah penyewaan sepeda harian".
- **Measurable (Terukur):** Dapat diukur dengan jelas menggunakan nilai rata-rata dari total penyewaan sepeda harian (kolom `cnt`) pada setiap kategori musim.
- **Action-Oriented (Berorientasi Aksi):** Hasil analisis ini akan membantu tim operasional untuk melakukan penyesuaian stok sepeda dan alokasi staf perawatan; misalnya memperbanyak armada di musim yang paling ramai dan melakukan *maintenance* besar-besaran di musim yang sepi.
- **Relevant (Relevan):** Sangat relevan untuk manajemen inventaris, efisiensi biaya, dan optimalisasi pendapatan bisnis penyewaan sepeda sepanjang tahun.
- **Time-bound (Terikat Waktu):** Terbatas pada rentang waktu data historis yang tersedia dalam dataset, yaitu sepanjang tahun 2011 hingga 2012.

**2. Pada jam berapa saja tingkat penyewaan sepeda mencapai puncaknya (rush hour) khusus pada hari kerja (workingday = 1)?**
- **Specific (Spesifik):** Fokus mencari waktu spesifik yaitu "jam sibuk (rush hour)" penyewaan yang hanya terjadi pada rentang "hari kerja" (Senin-Jumat).
- **Measurable (Terukur):** Dihitung secara konkret berdasarkan jumlah rata-rata penyewaan sepeda per jam (menggunakan kolom `cnt` pada jam atau kolom `hr` tertentu).
- **Action-Oriented (Berorientasi Aksi):** Memberikan arahan nyata bagi tim operasional untuk memastikan ketersediaan armada sepeda selalu penuh di stasiun-stasiun penyewaan tepat sebelum jam-jam sibuk tersebut dimulai.
- **Relevant (Relevan):** Sangat relevan dengan efisiensi operasional harian guna menghindari kekecewaan pelanggan (terutama kaum komuter) akibat kehabisan sepeda saat permintaan sedang sangat tinggi.
- **Time-bound (Terikat Waktu):** Dianalisis berdasarkan pola harian (dalam rentang 00:00 hingga 23:59) selama hari kerja operasional.

**3. Bagaimana perbandingan profil penyewa (Casual vs Registered) berdasarkan hari dalam seminggu? Siapa yang lebih mendominasi di hari kerja vs hari libur?**
- **Specific (Spesifik):** Membandingkan dua tipe pengguna secara spesifik, yaitu "Casual" (pengguna biasa/non-member) dan "Registered" (member terdaftar), serta menganalisis perbedaan perilaku mereka di "hari kerja" dibandingkan "akhir pekan/hari libur".
- **Measurable (Terukur):** Dapat diukur secara akurat dengan menjumlahkan dan membandingkan angka penyewaan pada kolom `casual` dan `registered` yang dikelompokkan berdasarkan hari (kolom `weekday`).
- **Action-Oriented (Berorientasi Aksi):** Hasilnya dapat digunakan oleh tim marketing untuk merancang kampanye promosi yang tepat sasaran (contoh: menawarkan promo *membership* di akhir pekan untuk mengubah pengguna *Casual* menjadi *Registered*).
- **Relevant (Relevan):** Sangat relevan dengan strategi segmentasi pelanggan, retensi pengguna, dan peningkatan jumlah pendaftaran keanggotaan bisnis penyewaan.
- **Time-bound (Terikat Waktu):** Pola perbandingan ini diamati dalam siklus mingguan (Senin hingga Minggu) menggunakan keseluruhan dataset yang ada.

**4. Bagaimana kondisi cuaca (weathersit) memengaruhi minat penyewa? Kondisi cuaca mana yang paling jarang terjadi penyewaan?**
- **Specific (Spesifik):** Fokus pada dampak dari berbagai "kondisi cuaca" (seperti cerah, mendung, hujan/salju ringan, cuaca ekstrem) terhadap "tingkat minat/jumlah penyewaan sepeda".
- **Measurable (Terukur):** Diukur menggunakan angka rata-rata penyewaan sepeda (kolom `cnt`) yang dikelompokkan secara jelas berdasarkan setiap kategori cuaca (kolom `weathersit`).
- **Action-Oriented (Berorientasi Aksi):** Membantu bisnis untuk melakukan langkah antisipasi operasional. Misalnya, mengembangkan sistem *warning* di aplikasi atau menawarkan voucher diskon khusus ketika cuaca sedang kurang bersahabat agar jumlah penyewaan tidak anjlok drastis.
- **Relevant (Relevan):** Sangat relevan dengan mitigasi risiko operasional, karena cuaca merupakan faktor penentu dan tantangan utama dalam bisnis transportasi luar ruangan (*outdoor*).
- **Time-bound (Terikat Waktu):** Berdasarkan evaluasi dari data riwayat cuaca harian dan per jam yang tercatat selama tahun 2011 dan 2012.
