## Data Wrangling

### Gathering Data
Memuat dataset `day.csv` dan `hour.csv` langsung dari repository GitHub menggunakan link Raw.

### Assessing Data
Menilai kualitas data untuk mencari tahu apakah terdapat *missing value*, duplikasi, atau kesalahan tipe data.

```python
print("--- Check Info & Missing Values ---")
day_df.info()
hour_df.info()

print("\n--- Check Duplicates ---")
print("Duplicate Day:", day_df.duplicated().sum())
print("Duplicate Hour:", hour_df.duplicated().sum())
```

#### Identifying Data Type Problem

**(Catatan: Tidak ada kode baru yang perlu dijalankan di sini karena kita sudah melihat masalahnya dari output `info()` di atas)**

**Steps to Take:**
- Mengubah tipe data pada kolom `dteday` di dataset `day_df` yang semula `object` (string) menjadi `datetime`.
- Mengubah tipe data pada kolom `dteday` di dataset `hour_df` yang semula `object` (string) menjadi `datetime`.

**Insight Assessing Data:**
- **Missing Value (Nilai Kosong):** Dataset ini sangat bersih. Tidak ditemukan adanya *missing value* pada `day_df` maupun `hour_df` (semua kolom memiliki jumlah *non-null* yang sama persis dengan total barisnya).
- **Duplikasi Data:** Tidak ada masalah duplikasi. Jumlah baris yang terduplikasi pada kedua dataset adalah 0.
- **Kesalahan Tipe Data:** Terdapat satu anomali tipe data, yaitu pada kolom `dteday` (tanggal). Saat ini kolom tersebut terdeteksi sebagai `object` (teks/string). Untuk mempermudah analisis berbasis waktu (*time-series*), tipe data ini wajib kita ubah menjadi `datetime` pada tahap *Cleaning Data* nanti.

### Cleaning Data
Pada tahap ini, kita akan memperbaiki masalah yang ditemukan pada tahap *Assessing Data*:
1. Mengubah tipe data kolom `dteday` dari `object` menjadi `datetime` pada kedua dataset.
2. Melakukan *mapping* pada kolom `season`, `weathersit`, dan `weekday` dari angka menjadi label teks agar lebih mudah dibaca saat divisualisasikan.
3. Mengubah tipe data `weekday` menjadi *Categorical* agar urutan harinya sesuai (Senin - Minggu) dan tidak berantakan saat di-plot.

#### Fixing ... problem

```python
# 1. Mengubah tipe data dteday menjadi datetime
day_df['dteday'] = pd.to_datetime(day_df['dteday'])
hour_df['dteday'] = pd.to_datetime(hour_df['dteday'])

# 2. Mapping label angka menjadi teks pada dataset day_df
day_df['season'] = day_df['season'].map({1: 'Spring', 2: 'Summer', 3: 'Fall', 4: 'Winter'})
day_df['weathersit'] = day_df['weathersit'].map({1: 'Clear', 2: 'Misty', 3: 'Light Snow/Rain', 4: 'Heavy Rain/Snow'})
day_df['weekday'] = day_df['weekday'].map({0: 'Sun', 1: 'Mon', 2: 'Tue', 3: 'Wed', 4: 'Thu', 5: 'Fri', 6: 'Sat'})

# 3. Mengatur urutan hari (Categorical Data)
days_order = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
day_df['weekday'] = pd.Categorical(day_df['weekday'], categories=days_order, ordered=True)

# Mengecek hasil cleaning
print("Info terbaru day_df:")
day_df.info()
display(day_df.head())
```

**Insight Cleaning Data:**
Setelah melakukan proses pembersihan dan modifikasi data, berikut adalah perubahan yang telah berhasil dilakukan:
1. **Perbaikan Tipe Data:** Kolom `dteday` pada `day_df` dan `hour_df` yang sebelumnya bertipe `object` (string) telah berhasil diubah menjadi tipe `datetime`. Hal ini akan sangat memudahkan kita jika ingin melakukan analisis berbasis waktu (*time-series*).
2. **Transformasi Label (Mapping):** Angka-angka pada kolom `season`, `weathersit`, dan `weekday` telah diubah menjadi teks kategori yang mudah dipahami (misalnya angka 1 pada musim menjadi 'Spring', angka 1 pada cuaca menjadi 'Clear', dll). Ini akan membuat visualisasi data nantinya menjadi jauh lebih informatif dan mudah dibaca oleh audiens.
3. **Pengurutan Kategori:** Kolom `weekday` telah diubah menjadi tipe data *Categorical* dengan urutan yang sudah ditetapkan (Mon, Tue, Wed, Thu, Fri, Sat, Sun) agar saat divisualisasikan, urutan harinya tidak berantakan sesuai abjad.

Saat ini, dataset sudah dalam keadaan sangat bersih, rapi, dan siap untuk tahap *Exploratory Data Analysis* (EDA)!

## Exploratory Data Analysis (EDA)




### 1. Eksplorasi Data Umum (General EDA)

#### B. EDA Univariate (Numerikal)

```python
# Histogram untuk variabel numerik
day_df[['temp', 'hum', 'windspeed', 'cnt']].hist(bins=30, figsize=(15, 10), color='skyblue', edgecolor='black')
plt.suptitle('Histogram Variabel Numerik')
plt.show()

# Boxplot untuk cek outlier pada jumlah penyewaan
plt.figure(figsize=(8, 5))
sns.boxplot(x=day_df['cnt'], color='lightcoral')
plt.title('Boxplot Jumlah Penyewaan Harian (Pengecekan Outlier)')
plt.show()
```

Artinya, semakin hangat cuacanya, semakin banyak orang yang menyewa sepeda. Sebaliknya, kelembaban (`hum`) dan kecepatan angin (`windspeed`) berkorelasi negatif (menurunkan minat penyewaan).

### 2. Eksplorasi Spesifik & Agregasi
Bagian ini dilakukan untuk menjawab pertanyaan bisnis yang telah ditetapkan sebelumnya melalui pengelompokan data (Agregasi).

#### Pertanyaan 1: Rata-rata Penyewaan Berdasarkan Musim (2011-2012)

```python
season_rent_df = day_df.groupby("season").agg({
    "cnt": ["mean", "max", "min", "std"]
})
display(season_rent_df)
```

**Insight:** - Rata-rata penyewaan sepeda tertinggi terjadi pada Musim Gugur (Fall / Musim 3), disusul oleh Musim Panas (Summer / Musim 2).
- Penyewaan paling rendah terjadi pada Musim Semi (Spring / Musim 1), yang kemungkinan disebabkan oleh cuaca yang masih cukup dingin atau belum stabil di awal tahun.

#### Pertanyaan 2: Pola Penyewaan Berdasarkan Jam pada Hari Kerja (Working Day)
Menggunakan `hour_df` untuk melihat jam puncak penyewaan.

```python
# Filter data hari kerja saja
workingday_hours = hour_df[hour_df["workingday"] == 1]

rush_hour_df = workingday_hours.groupby("hr").agg({
    "cnt": ["mean", "sum"]
}).reset_index()

display(rush_hour_df.sort_values(by=("cnt", "mean"), ascending=False).head(5))
```

**Insight:** - Pada hari kerja, lonjakan penyewaan sepeda (rush hour) terjadi sangat tajam pada pukul **17:00 dan 18:00** (sore hari) serta pukul **08:00** (pagi hari).
- Pola ini sangat identik dengan jam komuter, di mana orang-orang menyewa sepeda untuk berangkat bekerja di pagi hari dan pulang ke rumah di sore hari.

#### Pertanyaan 3: Perbandingan Pengguna Casual vs Registered Berdasarkan Hari

```python
user_type_weekday_df = day_df.groupby("weekday").agg({
    "casual": "mean",
    "registered": "mean",
    "cnt": "mean"
})
display(user_type_weekday_df)
```

**Insight:** - Pengguna **Registered** (terdaftar) secara konsisten mendominasi jumlah penyewaan setiap harinya, terutama pada rentang hari kerja (Senin - Jumat).
- Sebaliknya, pengguna **Casual** (biasa) menunjukkan tren yang unik: jumlah penyewaan mereka meningkat drastis pada akhir pekan (Sabtu dan Minggu). Hal ini mengindikasikan bahwa pengguna *casual* cenderung menyewa sepeda untuk rekreasi atau olahraga di hari libur.

#### Pertanyaan 4: Pengaruh Kondisi Cuaca terhadap Minat Penyewa

```python
weather_impact_df = day_df.groupby("weathersit").agg({
    "cnt": ["mean", "sum"]
})
display(weather_impact_df)
```

**Insight:** - Cuaca Cerah / Sedikit Berkabut (Weathersit 1) merupakan kondisi paling ideal dan mencatatkan rata-rata penyewaan tertinggi.
- Begitu cuaca memburuk menjadi Hujan/Salju Ringan (Weathersit 3), rata-rata penyewaan turun sangat drastis.
- Pada kondisi cuaca ekstrem atau buruk (Weathersit 4), aktivitas penyewaan sepeda hampir tidak ada sama sekali.

### Analisis Lanjutan (Advanced Analysis): Binning Variabel Suhu (Temperature)
Sebagai analisis tambahan, kita akan melakukan *binning* (pengelompokan) pada variabel numerik suhu (`temp`). Suhu akan dikategorikan ke dalam 3 rentang: **Dingin, Sedang, dan Panas**. Tujuannya adalah untuk melihat pola rata-rata penyewaan berdasarkan kategori suhu tersebut dengan cara yang lebih mudah dipahami.

```python
import pandas as pd

# Menggunakan pd.cut untuk binning suhu
bins = [0, 0.35, 0.70, 1.1]
labels = ['Dingin', 'Sedang', 'Panas']
day_df['temp_group'] = pd.cut(day_df['temp'], bins=bins, labels=labels, include_lowest=True)

# Melakukan agregasi melihat rata-rata penyewaan berdasarkan kategori suhu
temp_rentals = day_df.groupby('temp_group', observed=False)['cnt'].mean().reset_index()

print("=== Tabel Rata-rata Penyewaan Berdasarkan Kategori Suhu ===")
display(temp_rentals)
```

**Insight:** - Hasil *binning* menunjukkan konfirmasi yang jelas: rentang suhu **"Sedang"** dan **"Panas"** menghasilkan rata-rata penyewaan yang jauh lebih tinggi dibandingkan saat suhu sedang **"Dingin"**.
- Hal ini sejalan dengan hasil korelasi sebelumnya, membuktikan bahwa orang lebih suka bersepeda saat suhu lingkungan terasa hangat dan nyaman.

## Visualization & Explanatory Analysis

Bagian ini menyajikan visualisasi data untuk menjawab empat pertanyaan bisnis yang telah ditetapkan di awal proyek.

### Pertanyaan 1: Bagaimana pengaruh musim (season) terhadap rata-rata jumlah penyewaan sepeda harian selama tahun 2011 dan 2012?

**Explanatory Analysis:**

### Pertanyaan 2: Pada jam berapa saja tingkat penyewaan sepeda mencapai puncaknya (rush hour) khusus pada hari kerja?

```python
# Menghitung rata-rata penyewaan per jam pada hari kerja
workingday_hourly = hour_df[hour_df['workingday'] == 1].groupby('hr')['cnt'].mean().reset_index()

plt.figure(figsize=(12, 6))
sns.lineplot(x='hr', y='cnt', data=workingday_hourly, marker='o', color='tab:blue', linewidth=2)

plt.title('Pola Penyewaan Sepeda per Jam pada Hari Kerja', fontsize=15)
plt.xlabel('Jam (00-23)', fontsize=12)
plt.ylabel('Rata-rata Jumlah Penyewaan', fontsize=12)
plt.xticks(range(0, 24))
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

**Explanatory Analysis:**
Grafik garis di atas menunjukkan pola "dua puncak" (bimodal). Puncak pertama terjadi pada pagi hari sekitar pukul **08:00**, dan puncak kedua (yang paling tinggi) terjadi pada sore hari pukul **17:00 hingga 18:00**. Pola ini membuktikan bahwa pada hari kerja, sepeda digunakan sebagai alat transportasi utama untuk berangkat dan pulang kerja (komuter).

### Pertanyaan 3: Bagaimana perbandingan profil penyewa (Casual vs Registered) berdasarkan hari dalam seminggu?

```python
# Melakukan pivoting data untuk visualisasi perbandingan
user_type_comparison = day_df.groupby('weekday')[['casual', 'registered']].mean().reset_index()
user_type_melted = user_type_comparison.melt(id_vars='weekday', var_name='user_type', value_name='avg_rentals')

plt.figure(figsize=(12, 6))
sns.barplot(x='weekday', y='avg_rentals', hue='user_type', data=user_type_melted, palette='muted')

plt.title('Rata-rata Penyewaan: Casual vs Registered per Hari', fontsize=15)
plt.xlabel('Hari (0: Minggu, 6: Sabtu)', fontsize=12)
plt.ylabel('Rata-rata Jumlah Penyewaan', fontsize=12)
plt.legend(title='Tipe Pengguna')
plt.show()
```

**Explanatory Analysis:**
Analisis ini menunjukkan perbedaan perilaku yang signifikan:
1. **Registered Users:** Mendominasi penyewaan pada hari kerja (Senin-Jumat) dengan pola yang stabil.
2. **Casual Users:** Mengalami lonjakan yang sangat besar pada akhir pekan (Sabtu dan Minggu).
Dapat disimpulkan bahwa pengguna terdaftar adalah pekerja/pelajar (kebutuhan rutin), sedangkan pengguna casual adalah wisatawan atau warga yang berolahraga (kebutuhan rekreasi).

### Rekomendasi Bisnis (Action Items)

Berdasarkan kesimpulan di atas, berikut adalah beberapa rekomendasi aksi yang dapat dilakukan oleh perusahaan untuk mengoptimalkan operasional dan pendapatan:

**1. Strategi Pemasaran Berdasarkan Cuaca (Weather-Based Marketing)**
- Karena cuaca mendung/hujan ringan (Weathersit 3) menurunkan minat penyewaan secara drastis, tim marketing dapat memberikan **notifikasi promo atau diskon kilat (flash sale)** melalui aplikasi saat cuaca mulai memburuk. Ini dapat menjadi insentif bagi pengguna agar tetap mau menyewa sepeda meskipun cuaca kurang ideal.

**2. Optimalisasi Operasional di Jam Sibuk (Rush Hour Management)**
- Pada hari kerja (Senin-Jumat), tim operasional wajib melakukan redistribusi sepeda sebelum pukul **08:00 pagi** dan sebelum pukul **17:00 sore**. Pastikan stasiun sepeda di area perumahan penuh pada pagi hari, dan stasiun di area perkantoran/stasiun transit penuh pada sore hari untuk melayani para komuter (pengguna *Registered*).

**3. Program Konversi Pengguna Casual di Akhir Pekan**
- Karena pengguna *Casual* sangat aktif di akhir pekan (Sabtu-Minggu) untuk tujuan rekreasi, perusahaan bisa menawarkan program **"Weekend Pass"** atau promo khusus yang mengarahkan mereka untuk mendaftar menjadi member (*Registered*). Edukasi mereka mengenai keuntungan finansial jika menjadi member bulanan/tahunan.

**4. Penjadwalan Pemeliharaan (Maintenance Schedule)**
- Hindari melakukan perbaikan atau *maintenance* skala besar pada armada sepeda di Musim Gugur (Fall) karena itu adalah puncak penyewaan. Sebaiknya, alokasikan jadwal *maintenance* rutin dan besar-besaran pada **Musim Semi (Spring)** atau pada hari-hari di mana cuaca sedang sangat buruk (Weathersit 4), sehingga tidak ada potensi hilangnya pendapatan (lost sales) akibat sepeda yang sedang diperbaiki.
